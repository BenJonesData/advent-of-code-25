```python
def _is_invalid(i: int):
    i = str(i)
    l = len(i)
    if l % 2 == 1:
        return False
    elif i[:int(l/2)] == i[int(l/2):]:
        return True
    else: return False
    
  
def add_invalid_ids(input: str):
    id_pairs = [tuple(i.split("-")) for i in input.split(",")]

    output = 0
    for lower, upper in id_pairs:
        for i in range(int(lower), int(upper) + 1):
            if _is_invalid(i):
                output += i
    
    return output 
```

```python
def _is_invalid_new(i: int):
    i = str(i)
    l = len(i)
    
    for n in range(2, l + 1):
        if l % n == 0:
            is_valid = False
            block_size = int(l/n)
            sub = i[:block_size]
            for m in range(1, n):
                low = m * block_size
                hi = (m + 1) * block_size
                if i[low: hi] == sub:
                    continue
                else: 
                    is_valid = True
                    break
            
            if not is_valid:
                return True
    
    return False


def add_invalid_ids_new(input: str):
    id_pairs = [tuple(i.split("-")) for i in input.split(",")]

    output = 0
    for lower, upper in id_pairs:
        for i in range(int(lower), int(upper) + 1):
            if _is_invalid_new(i):
                output += i
    
    return output 
```

```python
test_solution_p2(add_invalid_ids_new)
```

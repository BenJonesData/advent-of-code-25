```python
def get_password(input):
    rotations_list = input.split("\n")

    zero_count = 0
    curr_line = 0
    curr_pos = 50
    for rot in rotations_list:
        curr_line += 1

        if len(rot) == 0:
            continue

        if rot[0] == "R":
            direction = 1
        elif rot[0] == "L":
            direction = -1
        else:
            raise ValueError(f"Invalid rotation direction on line {curr_line}")

        try:
            distance = int(rot[1:])
        except ValueError:
            raise ValueError(f"Invalid rotation distance on line {curr_line}")
        
        curr_pos += direction * distance
        curr_pos = curr_pos % 100

        if curr_pos == 0:
            zero_count += 1
    
    return zero_count
```

```python
test_solution_p1(get_password)
```

```python
input_1 = input_to_str('day_1_input.txt')

print(f"password: {get_password(input_1)}")
```

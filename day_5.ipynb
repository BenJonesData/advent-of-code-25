```python
def count_fresh(db):
    ranges, ids = [element.split("\n") for element in db.split("\n\n")]

    ranges = [rng.split("-") for rng in ranges]
    ranges = [(int(lo), int(hi)) for lo, hi in ranges]

    fresh_count = 0
    for i in ids:
        if i == "":
            continue
        for lo, hi in ranges:
            if lo <= int(i) <= hi:
                fresh_count += 1
                break
    
    return fresh_count
```

```python
def count_all_possible_fresh(db):
    ranges, _ = [element.split("\n") for element in db.split("\n\n")]
    ranges = [rng.split("-") for rng in ranges]
    ranges = [(int(lo), int(hi)) for lo, hi in ranges]
    ranges.sort(key=lambda x: x[0]) # AI assisted 

    count = 0

    num_ranges = len(ranges)
    current_range = 0
    done = False

    while not done:
        current_lo = ranges[current_range][0]
        current_hi = ranges[current_range][1]

        for i in range(current_range + 1, num_ranges):
            if ranges[i][0] > current_hi:
                current_range = i
                break

            if ranges[i][1] > current_hi:      
                current_hi = ranges[i][1]

            if i == num_ranges - 1:
                done = True
        
        count += 1 + current_hi - current_lo

    return count
```

```python
test_solution_p2(count_all_possible_fresh)
```

```python
count_all_possible_fresh(input)
```

### Part 1

```python
test_solution_p1 = create_test_function(TEST_INPUT, 21)
```

```python
TRANSITIONS = {1, -1}

def count_splits(diag):
    diag_list = [row for row in diag.split("\n") if row]

    current_cols = []
    current_cols.append(diag_list[0].index("S"))

    num_splits = 0
    for row in diag_list:
        if set(row) == {"."}:
            continue
        
        splitters_hit = []
        for col in current_cols:
            if row[col] == "^":
                splitters_hit.append(col)
                num_splits += 1
        
        for splitter in splitters_hit:
            current_cols.remove(splitter)
            for t in TRANSITIONS:
                new_col = splitter + t
                if new_col not in current_cols:
                    current_cols.append(new_col)

    return num_splits
```

```python
test_solution_p1(count_splits)
```

```python
count_splits(input)
```

### Part 2

```python
test_solution_p2 = create_test_function(TEST_INPUT, 40)
```

```python
def count_paths(diag):
    diag_list = [row for row in diag.split("\n") if row]
    cache = {}

    def count_from_point(row, col):
        if (row, col) in cache:
            return cache[(row, col)]

        if row >= len(diag_list):
            return 1
        
        full_row = diag_list[row]

        if full_row[col] == "^":
            count = 0
            for t in TRANSITIONS:
                count += count_from_point(row + 1, col + t)
        else:
            count = count_from_point(row + 1, col)

        cache[(row, col)] = count

        return count
    
    return count_from_point(0, diag_list[0].index("S"))
```

```python
test_solution_p2(count_paths)
```

```python
count_paths(input)
```

```python
def _find_largest_int_with_idx(int_str: str):
    largest = int(int_str[0])
    largest_idx = 0
    for idx, num in enumerate(int_str):
        num_int = int(num)
        if num_int > largest:
            largest = num_int
            largest_idx = idx
        
        if largest == 9:
            break
    
    return largest, largest_idx


def _largest_joltage(bank: str):
    first_num, first_num_idx = _find_largest_int_with_idx(bank[:-1])
    second_num, _ = _find_largest_int_with_idx(bank[first_num_idx + 1:])
    
    return 10 * first_num + second_num
```

```python
def add_largest_joltages(banks: str):
    banks_list = banks.split("\n")

    output = 0
    for bank in banks_list:
        if bank == "":
            continue
        output += _largest_joltage(bank)
    
    return output
```

```python
test_solution_p1(add_largest_joltages)
```

```python
add_largest_joltages(input)
```

```python
def _largest_joltage_12(bank: str):
    min_idx = 0
    joltage = 0
    for i in range(12):
        max_idx = len(bank) - 11 + i
        num, num_idx = _find_largest_int_with_idx(bank[min_idx: max_idx])
        min_idx += num_idx + 1

        joltage += num * (10 ** (11 - i))

    return joltage


def add_largest_joltages_12(banks: str):
    banks_list = banks.split("\n")

    output = 0
    for bank in banks_list:
        if bank == "":
            continue
        output += _largest_joltage_12(bank)
    
    return output
```

```python
test_solution_p2(add_largest_joltages_12)
```

```python
def product(ls: list[int]):
    product = 1
    for op in ls:
        product *= op
    return product


def add_solutions(input):
    input_clean = [[e for e in line.split(" ") if e] for line in input.split("\n") if line]
    num_probs = len(input_clean[-1])

    output = 0
    for i in range(num_probs):
        operands = [int(line[i]) for line in input_clean[:-1]]
        operator = input_clean[-1][i]
        if operator == "+":
            output += sum(operands)
        elif operator == "*":
            output += product(operands)
        else:
            raise ValueError(f"invalid operator on problem {i} - {operator}")
    
    return output
```

```python
test_solution_p1(add_solutions)
```

```python
add_solutions(input)
```

### Part 2

```python
test_solution_p2 = create_test_function(TEST_INPUT, 3263827)
```

```python
def add_new_solution(input):
    input_list = [line for line in input.split("\n") if line]
    line_length = len(input_list[0])
    max_number_length = len(input_list) - 1

    column_num = line_length - 1
    output = 0
    while True:

        operands = []
        while True:
            number = ""
            for i in range(max_number_length):
                entry = input_list[i][column_num]
                if entry != " ":
                    number += entry

            operands.append(int(number))

            if input_list[-1][column_num] != " ":
                operator = input_list[-1][column_num]
                break
            else:
                column_num -= 1

        if operator == "+":
            output += sum(operands)
        elif operator == "*":
            output += product(operands)
        
        if column_num == 0: 
            break
        else:
            column_num -= 2

    return output
```

```python
test_solution_p2(add_new_solution)
```

```python
add_new_solution(input)
```
